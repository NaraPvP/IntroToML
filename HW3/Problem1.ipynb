```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
breast_data.shape
```

```python
breast_input = pd.DataFrame(breast_data)
breast_input.head()
```

```python
breast_labels = breast.target
breast_labels.shape
```

```python
labels = np.reshape(breast_labels,(569,1))
final_breast_data = np.concatenate([breast_data,labels],axis=1)
final_breast_data.shape
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
features = breast.feature_names
features
```

```python
features_labels = np.append(features,'Type')
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['Type'].replace(0, 'Benign',inplace=True)
breast_dataset['Type'].replace(1, 'Malignant',inplace=True)
```

```python
breast_dataset.tail()
```

```python
#Problem 1
```

```python
model = GaussianNB()
```

```python
model.fit(breast.data, breast.target)
print(model)
```

```python
expected = breast.target
predicted = model.predict(breast.data)
```

```python
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
# Visualize confusion matrix using Heatmap
cnf_matrix = metrics.confusion_matrix(expected, predicted)

class_names = ['Benign', 'Malignant']
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("bottom")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
ax.set_xticklabels(['Benign', 'Malignant'])
ax.set_yticklabels(['Benign', 'Malignant'])
```

```python
# Compare with Logistic Regression
classifier = LogisticRegression(penalty='none',random_state=0, max_iter=10000)
LR_y = breast_dataset.pop('Type')
LR_x = breast_dataset.copy()
```

```python
classifier.fit(LR_x, LR_y)
```

```python
Y_pred = classifier.predict(LR_x)
```

```python
cnf_matrix = metrics.confusion_matrix(LR_y, Y_pred)
cnf_matrix
```
